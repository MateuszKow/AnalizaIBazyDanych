```python
#zaimportowanie potrzebnych pakietów
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#wczytanie oryginalnych danych
data=pd.read_csv("../Original Data/drinks.csv")
#wyświetlenie pierwszych 10 rekordów z bazy
data.head(10)
```

```python
#zmiana nazw kolumn
data=data.rename(columns={'country':'Country','beer_servings':'consumed cans of beer per-capita','spirit_servings':'consumed shots of spirits per-capita','wine_servings':'consumed glasses of wine per-capita','total_litres_of_pure_alcohol':'consumed litres of pure alcohol per-capita'})
#zapis danych ze zmienionymi nazwami kolumn
data.to_csv("../Analysis Data/drinks_with_new_colums_name.csv",index=False)
#wczytanie danych ze zmnienionymi nazwami kolumns
data=pd.read_csv("../Analysis Data/drinks_with_new_colums_name.csv")
#wyświetlenie danych
data
```

# Utworzenie listy 15 największych konsumentów piwa na świecie

```python
#posortowanie spożycia piwa w kolejności od największego konsumenta oraz pobranie 15 pierwszych rekordów
top_beer_consumers=(data.sort_values('consumed cans of beer per-capita',ascending=False)).head(15)
#zmiana indeksów
top_beer_consumers.index=[x for x in range(1,16)]
#wyświetlenie tylko danych o nazwie państwa oraz ilości spożytego piwa
top_beer_consumers=top_beer_consumers[['Country','consumed cans of beer per-capita']]
#zapis uzyskanych danych do pliku
top_beer_consumers.to_csv("../Analysis Data/top_beer_consumers.csv",index=False)
#wyświetlenie uzyskanych danych
top_beer_consumers
```

# Utworzenie listy 15 największych konsumentów napojów wysokoprocentowych na świecie

```python
#posortowanie spożycia alkoholi wysokoprocentowych w kolejności od największego konsumenta oraz pobranie 15 pierwszych rekordów
top_spirits_consumers=(data.sort_values('consumed shots of spirits per-capita',ascending=False)).head(15)
#zmiana indeksów
top_spirits_consumers.index=[x for x in range(1,16)]
#wyświetlenie tylko danych o nazwie państwa oraz ilości spożytych alkoholi wysokoprocentowych
top_spirits_consumers=top_spirits_consumers[['Country','consumed shots of spirits per-capita']]
#zapis uzyskanych danych do pliku
top_spirits_consumers.to_csv("../Analysis Data/top_spirits_consumers.csv",index=False)
#wyświetlenie uzyskanych danych
top_spirits_consumers
```

# Utworzenie listy 15 największych konsumentów wina na świecie

```python
#posortowanie spożycia wina w kolejności od największego konsumenta oraz pobranie 15 pierwszych rekordów
top_wine_consumers=(data.sort_values('consumed glasses of wine per-capita',ascending=False)).head(15)
#zmiana indeksów
top_wine_consumers.index=[x for x in range(1,16)]
#wyświetlenie tylko danych o nazwie państwa oraz ilości spożytego wina
top_wine_consumers=top_wine_consumers[['Country','consumed glasses of wine per-capita']]
#zapis uzyskanych danych do pliku
top_wine_consumers.to_csv("../Analysis Data/top_wine_consumers.csv",index=False)
#wyświetlenie uzyskanych danych
top_wine_consumers
```

# Utworzenie listy 15 największych konsumentów czystego alkoholu na świecie

```python
#posortowanie spożycia czystego alkoholu w kolejności od największego konsumenta oraz pobranie 15 pierwszych rekordów
top_alcohol_consumers=(data.sort_values('consumed litres of pure alcohol per-capita',ascending=False)).head(15)
#zmiana indeksów
top_alcohol_consumers.index=[x for x in range(1,16)]
#wyświetlenie tylko danych o nazwie państwa oraz ilości spożytego czystego alkoholu
top_alcohol_consumers=top_alcohol_consumers[['Country','consumed litres of pure alcohol per-capita']]
#zapis uzyskanych danych do pliku
top_alcohol_consumers.to_csv("../Analysis Data/top_alcohol_consumers.csv",index=False)
#wyświetlenie uzyskanych danych
top_alcohol_consumers
```

# Tworzenie wykresów prezentujących uzyskane dane

```python
#tworzenie wykresu
plt.figure(figsize=(16,8))
#wyrysowanie wykresu słupkowego największych konsumentów piwa
plt.bar(top_beer_consumers['Country'],top_beer_consumers['consumed cans of beer per-capita'],color=(0.2, 0.8, 0.6, 1))
#obrót etykiet osi X o 45 stopni
plt.xticks(rotation=45);
#nadanie tytułu wykresowi
plt.title("Top 15 biggest consumers of beer per-capita")
#opis osi Y
plt.ylabel("Consumed cans of beer per-capita")
#zapis wykresu do pliku o rozszerzeniu jpg
plt.savefig('../Analysis Data/top_beer_consumers.jpg')
#wyświetlenie wykresu
plt.show()
```

```python
#tworzenie wykresu
plt.figure(figsize=(16,8))
#wyrysowanie wykresu słupkowego największych konsumentów alkoholi wysokoprocentowych
plt.bar(top_spirits_consumers['Country'],top_spirits_consumers['consumed shots of spirits per-capita'],color=(0.2, 0.8, 0.1, 1))
#obrót etykiet osi X o 45 stopni
plt.xticks(rotation=45);
#nadanie tytułu wykresowi
plt.title("Top 15 biggest consumers of spirits per-capita")
#opis osi Y
plt.ylabel("Consumed shots of spirits per-capita")
#zapis wykresu do pliku o rozszerzeniu jpg
plt.savefig('../Analysis Data/top_spirits_consumers.jpg')
#wyświetlenie wykresu
plt.show()
```

```python
#tworzenie wykresu
plt.figure(figsize=(16,8))
#wyrysowanie wykresu słupkowego największych konsumentów wina
plt.bar(top_wine_consumers['Country'],top_wine_consumers['consumed glasses of wine per-capita'],color=(0.6, 0.2, 0.8, 1))
#obrót etykiet osi X o 45 stopni
plt.xticks(rotation=45);
#nadanie tytułu wykresowi
plt.title("Top 15 biggest consumers of wine per-capita")
#opis osi Y
plt.ylabel("Consumed glasses of wine per-capita")
#zapis wykresu do pliku o rozszerzeniu jpg
plt.savefig('../Analysis Data/top_wine_consumers.jpg')
#wyświetlenie wykresu
plt.show()
```

```python
#tworzenie wykresu
plt.figure(figsize=(16,8))
#wyrysowanie wykresu słupkowego największych konsumentów czystego alkoholu
plt.bar(top_alcohol_consumers['Country'],top_alcohol_consumers['consumed litres of pure alcohol per-capita'],color=(0.8, 0.8, 0.1, 0.6))
#obrót etykiet osi X o 45 stopni
plt.xticks(rotation=45);
#nadanie tytułu wykresowi
plt.title("Top 15 biggest consumers of pure alcohol per-capita")
#opis osi Y
plt.ylabel("Consumed litres of pure alcohol per-capita")
#zapis wykresu do pliku o rozszerzeniu jpg
plt.savefig('../Analysis Data/top_alcohol_consumers.jpg')
#wyświetlenie wykresu
plt.show()
```
